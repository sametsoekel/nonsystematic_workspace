## Facets and Multiplications 

```r
mpg %>% ggplot +
geom_point(aes(displ,hwy))+
facet_wrap(~class,nrow=4)
```

```r
mpg %>% ggplot() +
geom_point(aes(displ,hwy,color=drv)) + 
geom_smooth(aes(displ,hwy,color=drv))
```

```r
mpg %>% ggplot() +
geom_point(aes(displ,hwy,colour=drv)) + 
geom_smooth(aes(displ,hwy,group=drv),se=FALSE)
```
