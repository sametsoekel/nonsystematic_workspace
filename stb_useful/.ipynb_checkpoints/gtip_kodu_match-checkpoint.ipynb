```r
library(RecordLinkage)
library(data.table)
library(tidyverse)
```

```r
# RecordLinkage::levenshteinSim
```

```r
gtip <- fread('gtip_kod_oncelikli.csv')
yt <- fread('yatirim_tesvik_mayis_2020_kapasite.csv',header = F) %>% select(V1)
```

```r
match <- data.table()
for(i in yt[,.I]){
    levenshteinSim(str1 = yt[i,V1],str2=gtip[,gtip_tanimi]) %>% which.max -> index
    
    newrow <- data.table(kapasite=yt[i,V1],gtip_aciklama=gtip[index,gtip_tanimi])
    match <- match %>% rbind(newrow)
}
```
