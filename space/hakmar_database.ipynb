```r
library(RPostgres)
library(data.table)
library(readxl)
library(dplyr)
library(dbplyr)
options(warn=-1)
```

```r
drv <- dbDriver('Postgres')
con <- dbConnect(drv,user='samet',dbname='hakmar')
```

```r
setwd('/home/samet/Downloads/Hakmar Veri-20210405T192550Z-001/Hakmar Veri')
```

```r
cnames <- c()
for(i in 1:10) cnames[i] <- sprintf('V%s',i)
```

```r
for(i in c(2017,2018,2019,2020)){
    dum <- read_xlsx(sprintf('%s.xlsx',i),col_names = cnames)
    if(i==2017) {dbCreateTable(con,'veri',dum)} 
    setDT(dum)
    dum[,V1:=as.Date(V1)]
    
    dbWriteTable(con,'veri',dum,append=T,row.names=F)
}
```
