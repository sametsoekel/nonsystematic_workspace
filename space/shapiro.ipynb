```r
set.seed(1453)
options(repr.plot.width=15)

par(mfrow = c(1,2))

gaussian <- rnorm(200,0,1)
plot(density(gaussian),main=sprintf('shapiro-wilk p-val : %s',round(shapiro.test(gaussian)$p.value,4)))

uniform <- runif(200,0,1)
plot(density(uniform),main=sprintf('shapiro-wilk p-val : %s',round(shapiro.test(uniform)$p.value,4)))
```
